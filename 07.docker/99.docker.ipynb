1단계: 기존 패키지 업데이트

sudo apt-get update && sudo apt-get upgrade -y

2단계: 이전 Docker 버전 제거

sudo apt-get remove docker docker-engine docker.io containerd runc -y

3단계: 필수 구성 요소 설치

sudo apt install -y \
    ca-certificates \
    curl \
    gnupg \
    lsb-release

4단계: Docker의 공식 GPG 키 추가

sudo mkdir -p /etc/apt/keyrings
curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo gpg --dearmor -o /etc/apt/keyrings/docker.gpg

5단계: Docker 저장소 설정

echo \
  "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/docker.gpg] https://download.docker.com/linux/ubuntu \
  $(lsb_release -cs) stable" | sudo tee /etc/apt/sources.list.d/docker.list > /dev/null

6단계: 패키지 색인 다시 업데이트

sudo apt-get update

7단계: Docker 엔진, CLI 및 Containerd 설치

sudo apt-get install -y docker-ce docker-ce-cli containerd.io docker-buildx-plugin docker-compose-plugin

8단계: Docker 설치 확인

docker --version

9단계: Docker 서비스 시작 및 활성화

sudo systemctl start docker
sudo systemctl enable docker

10단계: sudo 없이 Docker 실행(선택 사항)

sudo usermod -aG docker $USER

변경 사항을 적용하려면 로그아웃했다가 다시 로그인하거나 터미널을 다시 시작.

docker run hello-world

#### 4-6-2-2 jenkins pipeline

```groovy
pipeline {
    agent any

    environment {
        PORT_ADMIN = '8090'
    }    

    stages {
        stage('Clone Repository') {
            steps {
                script {
                    // 클론할 디렉토리를 명시적으로 지정
                    // -d: 지정된 경로가 디렉토리인지 확인.
                    // [ ! -d "myproject" ] : "myproject"가 존재하지 않거나 디렉토리가 아닌 경우 참(true).                    
                    sh '''
                    if [ ! -d "myproject" ]; then
                        echo "myproject directory is not exist."
                    else
                    echo "Repository already cloned in myproject directory."
                        rm -rf ./myproject
                    fi
                    
                    git clone https://github.com/inwookbaek/20240809_green_cicd.git myproject
                    '''
                }
            }
        }

        stage('Check if App is Running') {
            steps {
                script {
                    sh '''
                    # Check if app.jar is running
                    if pgrep -f app.jar > /dev/null; then
                        # 프로세스 죽이기
                        pkill -9 -f app.jar
                    else
                        echo "app.jar is not running."
                    fi
                    '''
                }
            }
        }

        stage('Change Directory and Build') {
            steps {
                script {
                    // Change to the "myproject" directory and perform a task
                    sh '''
                    cd myproject
                    # pwd  # 확인용: 현재 디렉토리 출력
                    # ls   # 확인용: 디렉토리 내용 출력
                    
                    # ./gradlew 실행권한 부여
                    chmod +x ./gradlew
                    
                    # jar build
                    ./gradlew clean build
                    
                    # 필요한 빌드 명령 실행 (예: Docker 빌드, Gradle, Maven 등)
                    # 예: Docker 빌드
                    # docker build -t myimage:latest .
                    '''
                }
            }
        }  

        stage('Doker image Build') {
            steps {
                script {
                    // Change to the "myproject" directory and perform a task
                    sh '''
                    # 필요한 빌드 명령 실행 (예: Docker 빌드, Gradle, Maven 등)
                    # 예: Docker 빌드
                    # docker build -t spring-boot .
                    '''
                }
            }
        }  

        stage('Run Java App in Background') {
            steps {
                script {
                    sh '''
                    # Navigate to the project directory
                    cd myproject/build/libs
                    # pwd 
                    # ls
                    
                    # Run the Java application in the background
                    # nohup은 no hang up 의 약자 입니다. 해석 그대로 "끊지마!" 의미
                    # 중단 없이 실행하고자 하는 프로그램 명령어 앞에 "nohup" 만 붙여주면 된다.
                    # 참고 사이트 : https://joonyon.tistory.com/entry/쉽게-설명한-nohup-과-백그라운드-명령어-사용법
                    # 0 : 표준 입력/ 1 : 표준 출력 / 2 : 표준 에러
                    # 2>&1 : 표준출력과 표준에러를 같은 파일에 쓰기
                    export JENKINS_NODE_COOKIE=dontKillMe
                    nohup java -jar app.jar > app.log 2>&1 &
                    # nohup java -jar $WORKSPACE/myproject/build/libs/app.jar > $WORKSPACE/myproject/build/libs/app.log 2>&1 &
                    # nohup java -Dserver.port=${PORT_ADMIN} -Dspring.profiles.active=dev -Dfile.encoding=UTF-8 -jar app.jar > app.log 2>&1 &
                    
                    # Print confirmation
                    echo "Java application started in the background."
                    '''
                }
            }
        }
    }             
}
```

### github webhook과 jenkins 연동 at chatGPT
GitHub Webhook과 Jenkins를 연동하면 GitHub 저장소의 변경 사항(예: 코드 푸시, 풀 리퀘스트 등)이 발생할 때 Jenkins가 자동으로 빌드 파이프라인을 실행할 수 있습니다. 아래는 GitHub Webhook과 Jenkins를 연동하는 단계별 가이드입니다.

1. Jenkins에서 GitHub 플러그인 설치
Jenkins 관리 화면으로 이동: Jenkins Dashboard > Manage Jenkins > Manage Plugins.
Available 탭에서 GitHub Integration Plugin 또는 GitHub 관련 플러그인을 검색 후 설치.
설치 후 Jenkins를 재시작합니다.
2. Jenkins Job 설정
새로운 Freestyle 프로젝트 또는 Pipeline 프로젝트 생성:

Jenkins Dashboard > New Item.
Freestyle project 또는 Pipeline project 선택 후 이름 입력.
GitHub 저장소를 연동:

Freestyle 프로젝트:
Source Code Management에서 Git 선택.
저장소 URL과 자격 증명(예: Personal Access Token)을 추가.
Pipeline 프로젝트:
Pipeline 섹션에서 Git SCM 또는 Script에서 저장소를 클론하는 스크립트를 작성.
Build Triggers에서 "GitHub hook trigger for GITScm polling" 활성화.

3. GitHub 저장소에서 Webhook 설정
GitHub 저장소로 이동:
Settings > Webhooks 선택.
새로운 Webhook 생성:
클릭: Add webhook.
Webhook 구성:
Payload URL: Jenkins 서버의 URL 입력 (예: http://<jenkins-server>/github-webhook/).
로컬에서 Jenkins를 실행 중이라면, 외부에서 접근 가능하도록 ngrok 등을 사용.
Content type: application/json 선택.
Which events would you like to trigger this webhook?:
Just the push event 또는 필요한 이벤트 선택.
저장합니다.
4. Jenkins에서 GitHub Webhook 동작 테스트
GitHub 저장소에서 코드를 푸시하거나, Webhook Test 버튼을 사용하여 테스트.
Jenkins Job이 자동으로 트리거되는지 확인합니다:
Jenkins 대시보드에서 실행 중인 빌드 확인.
5. 추가: Jenkins Pipeline Script (예제)
GitHub Webhook을 활용한 Jenkins Pipeline 예제입니다.

## 1.1 docker command

### 1.1.1 상위커맨드 / 하위커맨드 /옵션
- docker container [start/stop/create/run/rm/exec/ls/cp/commit] [--name/-e/-p/-v/-d/-i/-t/-a]
- docker image     [pull/rm/ls/build] [-i]
- docker volumn    [create/inspect/ls/prune/rm]  [--name/-a]
- docker network   [connect/disconnect/create/inspect/ls/prune/rm] []

### 1.1.2 옵션

- --name 컨테이너이름
- -p 호스트번호:컨테이너포트번호      : -p = --publish
- -v 호스트디스크:컨테이너디렉토리    : -v = --volumn
- --net=네트워크명                    
- -e 환경변수이름=값                  : -e = --env
- -d : 백그라운드로 실행              : -d = --detach
- -i : 컨테이너에 터미널(키보드) 연결 : -i = --interactive
- -t : 특수키를 사용하도록 함         : -t = --tty
- -help : 도움말
