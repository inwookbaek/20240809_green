<div align="center">
  <img src="./images/02.aws_lightsail_ssh_03.jpg" width="800px" height="500px"><br>
  <div align="left">
    <ol>
      <li>다운로드된 폴더 외에서 실행할 경우 - 접속불가
        <div align="center">
          <img src="./images/02.aws_lightsail_ssh_04.jpg" width="800px" height="500px"><br>
          <img src="./images/02.aws_lightsail_ssh_05.jpg" width="800px" height="500px"><br>

        </div>
      </li> 
      <li>다운로드 폴더 안에서 실행할 경우, 공개키 or 새로운 개인키로 실행할 경우 둘다 실행이 않되고 원개인키로는 바로 실행가능
        <ul>
          <li>해결방법 aws-lightsail의 ssh로 접속후 실행</li> 
          <li>sudo vi /etc/ssh/sshd_config.d/*.conf</li> 
          <li>PasswordAuthentication no => PasswordAuthentication yes 로 변경</li> 
          <li>sudo systemctl restart ssh.service</li> 
          <li>exit</li> 
        </ul>
      </li> 
      <li>환경설정후 공개키 or 새로운 개인키 or 원개인키로 실행할 경우 모두 비번요구후 확인시 실행가능</li>
