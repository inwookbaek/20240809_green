````python
from youtube_transcript_api import YouTubeTranscriptApi

# 영상 ID
video_id = "FbMcz5vcUcA"

# 자막 추출
try:
    transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['ko'])  # 한국어 자막 추출
    for line in transcript:
        print(f"{line['text']} ({line['start']}초)")

    # 파일로 저장
    with open("transcript.txt", "w", encoding="utf-8") as file:
        for line in transcript:
            file.write(f"{line['text']} ({line['start']}초)\n")
    print("자막이 transcript.txt 파일로 저장되었습니다.")
except Exception as e:
    print(f"자막을 추출할 수 없습니다: {e}")
````

````python
from google.cloud import videointelligence_v1 as videointelligence

def analyze_video(video_uri):
    client = videointelligence.VideoIntelligenceServiceClient()
    features = [videointelligence.Feature.TEXT_DETECTION]

    operation = client.annotate_video(
        request={"features": features, "input_uri": video_uri}
    )

    print("\nProcessing video for text detection.")
    result = operation.result(timeout=90)
    print("\nFinished processing.\n")

    # 결과 출력
    for annotation in result.annotation_results[0].text_annotations:
        print(f"Text: {annotation.text}")
        for segment in annotation.segments:
            start_time = segment.segment.start_time_offset.total_seconds()
            end_time = segment.segment.end_time_offset.total_seconds()
            print(f"\tSegment: {start_time}s to {end_time}s")

# 동영상 분석 실행
video_uri = "gs://your-bucket-name/your-video-file.mp4"  # Google Cloud Storage에 업로드된 동영상
analyze_video(video_uri)
````

````python
import pytesseract
from PIL import Image
import os

# Windows의 경우, Tesseract 실행 파일 경로를 직접 지정
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# OCR 적용할 이미지 폴더
image_folder = "c:/temp/frames/"

# OCR 실행
for filename in sorted(os.listdir(image_folder)):
    if filename.endswith(".png"):
        img_path = os.path.join(image_folder, filename)
        text = pytesseract.image_to_string(Image.open(img_path))
        print(f"==== Code from {filename} ====\n{text}\n")
````

````python
# 3️⃣ GPT-4로 코드 오류 수정 및 보완

import openai

openai.api_key = "your_api_key"

def fix_code(code_text):
    prompt = f"Fix and complete this ReactJS or Spring Boot code:\n\n{code_text}"
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )
    return response["choices"][0]["message"]["content"]

# OCR로 추출한 코드 정리
extracted_code = """
const App = () => {
  return <MyComponnt />; // Incorrect component name
};
"""
fixed_code = fix_code(extracted_code)
print(fixed_code)
````
