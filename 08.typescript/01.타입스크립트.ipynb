```typescript
// 1. ë®¨ë²• : function*, yield*
function* gen() {
  yield* [1,2];
}

for(let value of gen()) {
  console.log(value);
}

// 2. ê¸°ë³¸ì˜ˆì œ
function* simpleGenerator() {
    console.log("ì²« ë²ˆì§¸ ì‹¤í–‰");
    yield 1; // ì²« ë²ˆì§¸ ì‹¤í–‰ì—ì„œ ë°˜í™˜ë˜ëŠ” ê°’
    
    console.log("ë‘ ë²ˆì§¸ ì‹¤í–‰");
    yield 2; // ë‘ ë²ˆì§¸ ì‹¤í–‰ì—ì„œ ë°˜í™˜ë˜ëŠ” ê°’
    
    console.log("ì„¸ ë²ˆì§¸ ì‹¤í–‰");
    yield 3; // ì„¸ ë²ˆì§¸ ì‹¤í–‰ì—ì„œ ë°˜í™˜ë˜ëŠ” ê°’

    console.log("ëª¨ë“  ì‘ì—… ì™„ë£Œ");
}

const gen = simpleGenerator(); // ìƒì„±ê¸° ê°ì²´ ìƒì„±

console.log(gen.next()); // { value: 1, done: false }
console.log(gen.next()); // { value: 2, done: false }
console.log(gen.next()); // { value: 3, done: false }
console.log(gen.next()); // { value: undefined, done: true }

// 3. ë¬´í•œ ë°˜ë³µ ìƒì„±ê¸° (nextë¡œ ê°’ ì „ë‹¬)
// ìƒì„±ê¸° í•¨ìˆ˜ì—ì„œ yieldë¡œ ì „ë‹¬ë°›ì€ ê°’ì„ ë‹¤ì‹œ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
function* counter() {
    let count = 1;
    while (true) {
        let increment = yield count; // next() í˜¸ì¶œ ì‹œ ì „ë‹¬ë°›ì€ ê°’ì„ ì €ì¥
        count += increment ?? 1; // ì „ë‹¬ëœ ê°’ì´ ì—†ìœ¼ë©´ ê¸°ë³¸ê°’ 1 ì¦ê°€
    }
}

const gen = counter();
console.log(gen.next()); // { value: 1, done: false }
console.log(gen.next(5)); // { value: 6, done: false }
console.log(gen.next(10)); // { value: 16, done: false }
console.log(gen.next()); // { value: 17, done: false }

// 4. for...ofë¡œ ìƒì„±ê¸° ë°˜ë³µí•˜ê¸°
// ìƒì„±ê¸°ëŠ” for...of ë¬¸ìœ¼ë¡œ ì‰½ê²Œ ë°˜ë³µí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
function* numbers() {
    yield 10;
    yield 20;
    yield 30;
}

for (let num of numbers()) {
    console.log(num);
}

// 5. ì‹¤ì „ ì˜ˆì œ: Fibonacci ìˆ˜ì—´ ìƒì„±ê¸°
function* fibonacci() {
    let a = 0, b = 1;
    while (true) {
        yield a;
        [a, b] = [b, a + b]; // í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´ ê³„ì‚°
    }
}

const fib = fibonacci();
console.log(fib.next().value); // 0
console.log(fib.next().value); // 1
console.log(fib.next().value); // 1
console.log(fib.next().value); // 2
console.log(fib.next().value); // 3
console.log(fib.next().value); // 5
```

```typescript
// âœ… 1. Promiseë€?
// ğŸ“Œ PromiseëŠ” ë¹„ë™ê¸° ì‘ì—…ì„ ìœ„í•œ ê°ì²´
// PromiseëŠ” ë¹„ë™ê¸° ì‘ì—…ì´ ì„±ê³µ(resolve)í•˜ê±°ë‚˜ ì‹¤íŒ¨(reject)í•˜ëŠ” ìƒíƒœë¥¼ ê´€ë¦¬í•˜ëŠ” ê°ì²´ì…ë‹ˆë‹¤.
// ì„¸ ê°€ì§€ ìƒíƒœê°€ ì¡´ì¬í•©ë‹ˆë‹¤.
// 1ï¸âƒ£ Pending (ëŒ€ê¸° ì¤‘): ì´ˆê¸° ìƒíƒœ, ì•„ì§ ì™„ë£Œë˜ì§€ ì•ŠìŒ.
// 2ï¸âƒ£ Fulfilled (ì´í–‰ë¨): ì‘ì—…ì´ ì„±ê³µì ìœ¼ë¡œ ì™„ë£Œë¨ (resolve).
// 3ï¸âƒ£ Rejected (ê±°ë¶€ë¨): ì‘ì—…ì´ ì‹¤íŒ¨í•¨ (reject).

// 1. ê¸°ë³¸ì˜ˆì œ
// async(function modifier)ë¥¼ ì‚¬ìš©í•œ í•¨ìˆ˜ëŠ” awaitë¥¼ ì‚¬ìš©ê°€ëŠ¥
// awaitëŠ” Promiseê°ì²´ë¥¼ í•´ì†Œ(resolve)í•œë‹¤.
async function get() {
  let values = []
  values.push(await Promise.resolve(1));
  values.push(await Promise.resolve(2));
  values.push(await Promise.resolve(3));
  return values;
}

get().then(values => console.log(values));

// âœ… 2. ğŸ“Œ Promise ê¸°ë³¸ ì‚¬ìš©ë²•
function asyncTask() {
    return new Promise((resolve, reject) => {
        setTimeout(() => {
            const success = Math.random() > 0.5; // 50% í™•ë¥ ë¡œ ì„±ê³µ ë˜ëŠ” ì‹¤íŒ¨
            if (success) {
                resolve("âœ… ì‘ì—… ì„±ê³µ!");
            } else {
                reject("âŒ ì‘ì—… ì‹¤íŒ¨!");
            }
        }, 2000);
    });
}

asyncTask()
    .then(result => console.log(result))  // ì„±ê³µ ì‹œ ì‹¤í–‰
    .catch(error => console.error(error)) // ì‹¤íŒ¨ ì‹œ ì‹¤í–‰
    .finally(() => console.log("ğŸ¯ ì‘ì—… ì™„ë£Œ!")); // í•­ìƒ ì‹¤í–‰

// ğŸ”¹ ì‹¤í–‰ ê²°ê³¼ (ëœë¤)

// âœ… 3. async/awaitë€?
// ğŸ“Œ async/awaitëŠ” Promiseë¥¼ ë” ì‰½ê²Œ ì‚¬ìš©í•˜ê¸° ìœ„í•œ ë¬¸ë²•
// async í•¨ìˆ˜ëŠ” í•­ìƒ Promiseë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
// await í‚¤ì›Œë“œëŠ” ë¹„ë™ê¸° ì‘ì—…ì´ ëë‚  ë•Œê¹Œì§€ ê¸°ë‹¤ë¦½ë‹ˆë‹¤.
// try...catchë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜¤ë¥˜ë¥¼ ì‰½ê²Œ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

// ğŸ“Œ async/await ê¸°ë³¸ ì‚¬ìš©ë²•
async function fetchData() {
    try {
        console.log("â³ ë°ì´í„° ê°€ì ¸ì˜¤ëŠ” ì¤‘...");
        let response = await asyncTask();
        console.log(response);
    } catch (error) {
        console.error(error);
    } finally {
        console.log("ğŸ¯ ì‘ì—… ì™„ë£Œ!");
    }
}

fetchData();

// ğŸ”¹ ì‹¤í–‰ ê²°ê³¼ (ëœë¤)

// âœ… 4. Promise.all()ì„ ì‚¬ìš©í•œ ë³‘ë ¬ ì²˜ë¦¬
// ë¹„ë™ê¸° ì‘ì—…ì„ ë™ì‹œì— ì‹¤í–‰í•˜ê³ , ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë  ë•Œê¹Œì§€ ê¸°ë‹¤ë¦¬ê³  ì‹¶ì„ ë•Œ // Promise.all()ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

function task1() {
    return new Promise(resolve => setTimeout(() => resolve("ğŸ“Œ Task 1 ì™„ë£Œ!"), 2000));
}
function task2() {
    return new Promise(resolve => setTimeout(() => resolve("ğŸ“Œ Task 2 ì™„ë£Œ!"), 3000));
}
function task3() {
    return new Promise(resolve => setTimeout(() => resolve("ğŸ“Œ Task 3 ì™„ë£Œ!"), 1000));
}

async function runTasks() {
    console.log("ğŸš€ ì‘ì—… ì‹œì‘...");
    const results = await Promise.all([task1(), task2(), task3()]);
    console.log(results);
    console.log("ğŸ¯ ëª¨ë“  ì‘ì—… ì™„ë£Œ!");
}

runTasks();

// âœ… 5. Promise.race()ë¥¼ ì‚¬ìš©í•œ ê°€ì¥ ë¹ ë¥¸ ì‘ì—… ë°˜í™˜
// Promise.race()ëŠ” ê°€ì¥ ë¹¨ë¦¬ ëë‚˜ëŠ” ë¹„ë™ê¸° ì‘ì—… í•˜ë‚˜ë§Œ ë°˜í™˜í•©ë‹ˆë‹¤.

async function fastestTask() {
  const result = await Promise.race([task1(), task2(), task3()]);
  console.log("ğŸš€ ê°€ì¥ ë¹¨ë¦¬ ëë‚œ ì‘ì—…:", result);
}

fastestTask();

// âœ… 6. Promise.allSettled()ë¥¼ ì‚¬ìš©í•œ ëª¨ë“  ì‘ì—… ê²°ê³¼ í™•ì¸
// Promise.allSettled()ëŠ” ëª¨ë“  ì‘ì—…ì˜ ì„±ê³µ/ì‹¤íŒ¨ ì—¬ë¶€ë¥¼ ê°œë³„ì ìœ¼ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.

function failingTask() {
    return new Promise((_, reject) => setTimeout(() => reject("âŒ ì‹¤íŒ¨í•œ ì‘ì—…!"), 1500));
}

async function checkAllTasks() {
    const results = await Promise.allSettled([task1(), failingTask(), task3()]);
    console.log(results);
}

checkAllTasks();

// âœ… 7. async/awaitì„ í™œìš©í•œ API í˜¸ì¶œ ì˜ˆì œ
async function getUser() {
    try {
        let response = await fetch("https://jsonplaceholder.typicode.com/users/1");
        let data = await response.json();
        console.log("ğŸ‘¤ ì‚¬ìš©ì ì •ë³´:", data);
    } catch (error) {
        console.error("âŒ API í˜¸ì¶œ ì‹¤íŒ¨:", error);
    }
}

getUser();
```
